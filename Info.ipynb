```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds=pd.read_csv('./data/total_cases.csv')
data=np.array(ds)
```

```python
sns.barplot(data[:14,0][::-1],data[:14,1][::-1],palette="RdBu_r")
plt.ylabel('Female foeticide Cases')
plt.xlabel('Year')
plt.title('Female foeticide Cases 2001-2014')
plt.show()
```

```python
table=pd.read_csv('foeticie.csv')
```

```python
test=table[['States']]
test['2011-2014']=table['2011']+table['2012']+table['2013']+table['2014']
test=np.array(test).tolist()
test.sort(key=lambda x:x[1])
x=[k[0] for k in test][::-1]
y=[k[1] for k in test][::-1]
sns.set(font_scale=1) 
f, ax = plt.subplots(figsize=(10, 10))
#sns.barplot(table['2014'],table['States'],palette="Blues_d")
sns.barplot(y,x,palette="RdBu_r")
plt.ylabel('STATES')
plt.xlabel('NUMBER OF CASES')
plt.title('Foeticide Cases:2011-2014')
plt.show()
```

```python
literacy=pd.read_csv('literacy.csv')
```

```python
census=pd.read_csv('census.csv')
```

```python
table_sex_ratio=census[['States','SexRatio']]
table_sex_ratio=np.array(table_sex_ratio).tolist()
table_sex_ratio.sort(key=lambda x:x[1])
tx=[a[0] for a in table_sex_ratio][:20]
ty=[a[1] for a in table_sex_ratio][:20]
#sns.set(font_scale=1) 
#f, axes = plt.subplots(2,figsize=(5,5))
f, (ax1, ax2) = plt.subplots(1,2, figsize=(10, 5))
sns.barplot(ty,tx,palette="RdBu_r",ax=ax1)
ax1.set_xlabel('Sex Ratio')
ax1.set_ylabel('Bottom 20 States')
sns.barplot(y[:13],x[:13],palette="RdBu_r",ax=ax2)
ax2.set_xlabel('No. Cases 2011-2014')
#plt.setp(axes)
#plt.title('Sex Ration Vs No. of cases')
plt.tight_layout()
plt.show()
```

## Comparing Sex Ratio with Number of Foecitide Cases
  States with lower sex ratio tends to be where the foecitide cases are more common

```python
table_literacy=census[['States','Literacy']]
table_literacy=np.array(table_literacy).tolist()
table_literacy.sort(key=lambda x:x[1])
f, (ax1, ax2) = plt.subplots(1,2, figsize=(10, 5))
lit_x=[c[0] for c in table_literacy][:20]
lit_y=[c[1] for c in table_literacy][:20]
sns.barplot(lit_y,lit_x,palette="RdBu_r",ax=ax1)
ax1.set_xlabel('Literacy Rate')
ax1.set_ylabel('Bottom 20 States')
sns.barplot(y[:13],x[:13],palette="RdBu_r",ax=ax2)
ax2.set_xlabel('No. Cases 2011-2014')
plt.tight_layout()
plt.show()
```

## Comparing Literacy rate Vs No. of foeticide Cases
 States with low literacy rate tends to be where the foecitide cases are more common
