```python
from bs4 import BeautifulSoup
import urllib
r = urllib.urlopen('http://www.census2011.co.in/states.php').read()
soup = BeautifulSoup(r)
```

```python
print soup
```

```python
a=soup.find_all("tbody",class_="searchable")[0]
```

```python
states,pop,area,density,sex_ratio,literacy=[],[],[],[],[],[]
```

```python
b=a.find_all('td')[8:]
k=0
for i in range(35):
    x=str(b[k+2].get_text())
    x= int(x.replace(',', ''))
    y=str(b[k+4].get_text())
    y= int(y.replace(',', ''))
    z=str(b[k+5].get_text())
    z= int(z.replace(',', ''))
    v=str(b[k+6].get_text())
    v= int(v.replace(',', ''))
    states.append(str(b[k+1].get_text()))
    pop.append(x)
    area.append(y)
    density.append(z)
    sex_ratio.append(v)
    literacy.append(float(b[k+7].get_text()))
    k+=8
```

```python
import pandas as pd
```

```python
census=pd.DataFrame({'States':states,'Population':pop,'Area':area,'Density':density,'SexRatio':sex_ratio,'Literacy':literacy})
```

```python
census.to_csv('census.csv',header=True, columns=['States', 'Population','Area','Density','SexRatio','Literacy'], sep=',', index=False)
```
